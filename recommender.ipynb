```python
import numpy as np
import pandas as pd
```

```python
movies= pd.read_csv('tmdb_5000_movies.csv.zip')
credits= pd.read_csv('tmdb_5000_credits.csv.zip')
```

```python
movies.head(1)
```

```python
credits.head()
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
#keep columns
#genres id keywords title overview cast crew
movies=movies[['movie_id','cast','crew','genres','keywords','overview','title']]
```

```python
movies.head(1)
```

DATA PREPROCESSING


```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
#check duplicates
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
  l=[]
  for i in ast.literal_eval(obj):
    l.append(i['name'])
  return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.head(1)
```

```python
movies.iloc[0].keywords
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
movies.iloc[0].cast
```

```python
#top 3 cast
import ast
def convert2(obj):
  l=[]
  c=0
  for i in ast.literal_eval(obj):
    if c!=3:
      l.append(i['name'])
      c+=1
    else:
      break
  return l
```

```python
movies['cast']=movies['cast'].apply(convert2)
```

```python
movies.head()
```

```python
movies.iloc[0].crew
```

```python
#crew name whose job is director
import ast
def getdirector(obj):
  l=[]
  c=0
  for i in ast.literal_eval(obj):
    if i['job']=='Director':
      l.append(i['name'])
      break
  return l
```

```python
movies['crew']=movies['crew'].apply(getdirector)
```

```python
movies.head(1)
```

```python
movies.iloc[0].overview
```

```python
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres']=movies['genres'].apply(lambda x: [i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x: [i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x: [i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x: [i.replace(" ","") for i in x])
```

```python
movies.head(1)
```

```python
movies['tag']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
movies.head()
```

```python
new= movies[['movie_id','title','tag']]
```

```python
new.head(1)
```

```python
new['tag']=new['tag'].apply(lambda x:" ".join(x))
```

```python
new.head(1)
```

```python
new['tag'][0]
```

```python
new['tag']=new['tag'].apply(lambda x:x.lower())
```

```python
new.head(1)
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
  l=[]
  for i in text.split():
      l.append(ps.stem(i))
  return " ".join(l)
```

```python
new['tag']=new['tag'].apply(stem)
```

VECTORIZATION OF TAGS

```python
#vectorization using bag of words technique
#in bag of words all tags are merged together and the top n words which appear most frequently are fetched. then each tag is checked for all those words and a vector for each tag is created on the basis of the frequency of those words.
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors=cv.fit_transform(new['tag']).toarray()
vectors[0]
```

```python
for i in cv.get_feature_names_out():
  print(i)
```

```python
#use cosine distance to measure the similarity of each vector with one another
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#similarity of each movie with all the other movies
similarity=cosine_similarity(vectors)
```

```python
#similarity of first movie with other movies
#keeping track of the indexes so after sorting the index won't be lost
#reverse on the basis of second value not index in descending order
sorted(list(enumerate(similarity[1])),reverse=True,key=lambda x:x[1])[1:6]
```

```python
def recommend(movie):
  #fetch index of movie
  movie_index=new[new['title']==movie].index[0]
  #fetching similarity of movie with other movies from similarity matrix
  distance=similarity[movie_index]
  movie_recommend=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1])[1:6]
  for i in movie_recommend:
      print(new.iloc[i[0]].title)
```
